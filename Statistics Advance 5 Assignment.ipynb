```python
#Q1
import scipy.stats as stats

# Sample statistics
sample_mean = 50
sample_std_dev = 5

# Confidence level (95%)
confidence_level = 0.95

# Sample size (assuming a sufficiently large sample for the normal approximation)
sample_size = 100  # Adjust this if you have the actual sample size

# Calculate the standard error of the mean
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error using the z-score for the desired confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Two-tailed test
margin_of_error = z_score * standard_error

# Calculate the confidence interval
lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

# Print the results
print(f"95% Confidence Interval: ({lower_limit:.2f}, {upper_limit:.2f})")
```

```python
Interpretation:
The 95% confidence interval for the population mean based on the given sample data is calculated to be 
between approximately 47.96 and 52.04. This means that we can be 95% confident that the true population
mean falls within this interval.

Keep in mind that the confidence interval assumes that the data follows a normal distribution and that 
the sample size is sufficiently large for the normal approximation. Adjust the sample_size variable if
you have the actual sample size.
```

```python
#Q2
import numpy as np
from scipy.stats import chi2_contingency

# Observed frequencies of each color
observed_frequencies = np.array([35, 15, 30, 10, 12, 18])  # Replace with your actual data

# Expected frequencies based on the expected distribution
expected_frequencies = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * observed_frequencies.sum()

# Perform the chi-square goodness of fit test
chi2, p, dof, expected = chi2_contingency(observed_frequencies.reshape(1, -1))

# Define the significance level
alpha = 0.05

# Print the test statistics and p-value
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Compare p-value to significance level
if p < alpha:
    print("Reject the null hypothesis: The distribution does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution matches the expected distribution.")
```

```python
#Q4
import scipy.stats as stats
import numpy as np

# Sample size and number of individuals who smoke
n = 500
x = 60

# Calculate the sample proportion
p_hat = x / n

# Set the confidence level (alpha) and construct the confidence interval
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)  # Z-score for a 95% confidence interval

# Calculate the margin of error
margin_of_error = z_critical * np.sqrt(p_hat * (1 - p_hat) / n)

# Calculate the lower and upper bounds of the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

# Print the confidence interval
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

```

```python
#Q6
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Generate x values for the chi-square distribution
x = np.linspace(0, 30, 1000)

# Calculate the chi-square probability density function (PDF) for the given degrees of freedom
pdf = stats.chi2.pdf(x, df)

# Create a plot
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label=f'Chi-Square (df={df})', color='blue')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
pdf_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, pdf_fill, color='red', alpha=0.5, label='Chi-Square Statistic = 15')

# Label axes and add legend
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#Q8
To conduct a chi-square goodness of fit test to determine if the observed frequencies match the 
expected frequencies of a fair coin, you can follow these steps:

Step 1: State the Hypotheses

Null Hypothesis (H0): The coin is fair, and the observed frequencies match the expected frequencies.
Alternative Hypothesis (Ha): The coin is biased, and the observed frequencies do not match the expected
frequencies.

Step 2: Set the Significance Level

Set the significance level () to 0.05.

Step 3: Define the Expected Frequencies

For a fair coin, the expected frequencies for heads (H) and tails (T) in 100 coin flips are both 50.

Step 4: Calculate the Chi-Square Test Statistic

The chi-square test statistic can be calculated using the formula:

^2 =  [(Observed Frequency - Expected Frequency)^2 / Expected Frequency]

Calculate this for both heads (H) and tails (T):

For Tails (T):
^2_T = [(45 - 50)^2 / 50] = 5^2 / 50 = 25 / 50 = 0.5

For Heads (H):
^2_H = [(55 - 50)^2 / 50] = 5^2 / 50 = 25 / 50 = 0.5

Step 5: Determine the Degrees of Freedom

Degrees of freedom (df) = Number of Categories - 1
In this case, df = 2 - 1 = 1

Step 6: Find the Critical Value

Using a chi-square distribution table or calculator with df = 1 and  = 0.05, you find the critical
chi-square value is approximately 3.841.

Step 7: Compare the Calculated Chi-Square Statistic

Compare the calculated chi-square statistic (^2) to the critical chi-square value:

If ^2 < 3.841, fail to reject the null hypothesis (H0).
If ^2  3.841, reject the null hypothesis (H0) in favor of the alternative hypothesis (Ha).
In this case, ^2_T = ^2_H = 0.5, which is less than 3.841.

Step 8: Make a Decision

Since ^2 < 3.841 for both Tails and Heads, we fail to reject the null hypothesis. There is not enough
evidence to conclude that the coin is biased towards tails.

Step 9: Conclusion

Based on the chi-square goodness of fit test at a significance level of 0.05, we do not have sufficient 
evidence to support the hypothesis that the coin is biased towards tails. The coin appears to be fair.
```
